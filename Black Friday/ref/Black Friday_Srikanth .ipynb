```python
x=df.iloc[:,:-1]
```

```python
x.columns
```

```python
y=df['Purchase']
```

```python
from catboost import CatBoostRegressor
```

```python
cat = []
p=0
for i in df.dtypes:
    if(i == 'object'):
        cat.append(p)
    p+=1
```

```python
x.dtypes
```

```python
from catboost import CatBoostRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=24)
```

```python
model = CatBoostRegressor(iterations=300,loss_function='RMSE')
model.fit(x_train,y_train,cat_features=cat,eval_set = (x_test,y_test), plot=True)
```

### Though 250 trees gave us good results we will consider only 150 for hyper parameter tuning because of time factor

```python
from sklearn.model_selection import GridSearchCV
```

```python
param1 ={
    'depth' : range(1,16,2)
}
grid1 = GridSearchCV(estimator = CatBoostRegressor(iterations=150,loss_function='RMSE',random_seed=24,cat_features=cat),param_grid=param1,cv=5,scoring='neg_mean_squared_error')
```
